```python
def parse_revenue_range(value):
    match = re.findall(r'(\d+\.?\d*)', value)
    if len(match) == 2:
        low, high = map(float, match)
    else:
        return pd.Series([None, None])
    if 'billion' in value:
        low *= 1e9
        high *= 1e9
    elif 'million' in value:
        low *= 1e6
        high *= 1e6
    return pd.Series([low, high])
```

```python
df[['Revenue_min', 'Revenue_max']] = df['Revenue'].apply(parse_revenue_range)
```

```python
df['Avg_Rev']=(df['Revenue_min']+df['Revenue_max'])/2
```

```python
def parse_size_range(value):
    if pd.isna(value):
        return np.nan, np.nan  # handle missing values
    value = str(value)

    # find numbers (like 201, 500, 10000, etc.)
    match = re.findall(r'\d+', value)

    if 'to' in value and len(match) == 2:
        # example: "201 to 500 employees"
        low, high = map(int, match)
    elif '+' in value and len(match) == 1:
        # example: "10000+ employees"
        low = int(match[0])
        high = np.inf  # means "more than"
    elif len(match) == 1:
        # example: "500 employees"
        low = high = int(match[0])
    else:
        low = high = np.nan

    return low, high

# apply to your column
df[['Size_min', 'Size_max']] = df['Size'].apply(lambda x: pd.Series(parse_size_range(x)))
```

```python
df['Avg_Size']=(df['Size_min']+df['Size_max'])/2
```
